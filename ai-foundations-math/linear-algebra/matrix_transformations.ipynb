```python
import numpy as np
import matplotlib.pyplot as plt

def plot_transform(matrix, vectors, title):
    plt.figure(figsize=(6, 6))
    colors = ['r', 'b', 'g']
    for i, vec in enumerate(vectors):
        transformed = matrix @ vec
        plt.quiver(0, 0, vec[0], vec[1], angles='xy', scale_units='xy', scale=1, color=colors[i], label=f'Original {i+1}')
        plt.quiver(0, 0, transformed[0], transformed[1], angles='xy', scale_units='xy', scale=1, linestyle='dotted', color=colors[i], label=f'Transformed {i+1}')
    plt.xlim(-4, 4)
    plt.ylim(-4, 4)
    plt.grid()
    plt.axhline(0, color='black')
    plt.axvline(0, color='black')
    plt.gca().set_aspect('equal')
    plt.title(title)
    plt.legend()
    plt.show()

# Example: rotation matrix (90 degrees)
theta = np.pi / 2
rotation = np.array([[np.cos(theta), -np.sin(theta)],
                     [np.sin(theta),  np.cos(theta)]])
v1 = np.array([1, 0])
v2 = np.array([0, 1])
plot_transform(rotation, [v1, v2], "90° Rotation")
```
