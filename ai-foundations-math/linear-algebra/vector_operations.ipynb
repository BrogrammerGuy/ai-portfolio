```python
# This script demonstrates vector operations such as addition and scalar multiplication,
# and visualizes the results using a reusable plotting function.


# Define a reusable function to plot vectors
# This function can be used to plot any number of vectors with specified colors and labels.
def plot_vectors(vectors, colors=None, labels=None, title=None, xlim=(-5, 7), ylim=(-5, 7)):
    import matplotlib.pyplot as plt
    plt.figure(figsize=(6, 6))
    if colors is None:
        colors = ['r', 'b', 'g', 'm', 'c', 'y', 'k']
    for i, vec in enumerate(vectors):
        color = colors[i % len(colors)]
        label = labels[i] if labels and i < len(labels) else f'vector {i+1}'
        plt.quiver(0, 0, vec[0], vec[1], angles='xy', scale_units='xy', scale=1, color=color, label=label)
    plt.xlim(*xlim)
    plt.ylim(*ylim)
    plt.grid()
    plt.axhline(0, color='black')
    plt.axvline(0, color='black')
    plt.gca().set_aspect('equal')
    if title:
        plt.title(title)
    plt.legend()
    plt.show()


# Main logic to perform vector operations and plot the results

# Import necessary libraries
import numpy as np

# Define two vectors
v1 = np.array([2, 1])
v2 = np.array([1, 3])

# Vector addition
v3 = v1 + v2

# Scalar multiplication
v4 = -0.5 * v3

# Dot product of v1 and v2
dot_product = np.dot(v1, v2)
v5 = np.array([dot_product, 0])  # Create a vector for the dot product result = [5, 0]

# Plot all vectors using the reusable function
plot_vectors(
    [v1, v2, v3, v4, v5],
    colors=['r', 'b', 'g', 'magenta'],
    labels=['v1', 'v2', 'v1 + v2', '-0.5 * (v1 + v2)'],
    title="Vector Operations"
)
```
